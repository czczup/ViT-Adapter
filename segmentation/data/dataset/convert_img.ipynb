```python
from PIL import Image
import os
import numpy as np
names = []
path = 'annotations/training'
#path = 'annotations/validation'
#path = 'images/training'
#path = 'images/validation'
#path = '../ade/ADEChallengeData2016/annotations/validation'
for img in os.listdir(path):
    names.append(img)
names.sort()
print(len(names))
```

```python
from torchvision.transforms.functional import rotate
count=0
imgs=[]
for i,name in enumerate(names):
    img = Image.open(path + "/" + name)#.convert('L')
    imgs.append(np.array(img))
    rotate(img,90).save(path + "/" + name[:-4] + '_90.jpg', quality=100, subsampling=0)
    rotate(img,180).save(path + "/" + name[:-4] + '_180.jpg', quality=100, subsampling=0)
    rotate(img,270).save(path + "/" + name[:-4] + '_270.jpg', quality=100, subsampling=0)

```
